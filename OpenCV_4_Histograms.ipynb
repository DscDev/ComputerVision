```python
# los Histogramas son una manera increible de vizualizar componentes
# del color por individual

import cv2
import numpy as np
# necesitamos importar matplotlib para crear nuestros graficos de 
# histograma
from matplotlib import pyplot as plt

image = cv2.imread('images/input.png')
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# graficamos un histograma, ravel() aplana nuestro array de la imagen
plt.hist(image.ravel(), 256, [0, 256]); 
plt.show()

# viendo separados canales de colores
color = ('b', 'g', 'r')

# Ahora separamos los colores y graficamos cada uno en el Histograma
for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histogram2, color = col)
    plt.xlim([0,256])

    plt.show()
```
